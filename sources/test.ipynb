```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

```python
img1= cv2.imread('../Frames_channels/ENTV.jpg')
img1= cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img1 = cv2.resize(img1,(360,180))
```

```python
def match_img(img1,img2):
    match_score = np.mean(img2==img1)
    if match_score > .90:
        print('match' ,match_score)
    else:
        print('no match',match_score)
```

```python
cap =cv2.VideoCapture('/Users/anis/PycharmProjects/TV-Advertisements-Detection/Recordings/output_0_0:00:27_2021-10-03 23:25:00.566857.mp4')
while True :
    ret, current_frame = cap.read()
    current_frame= cv2.cvtColor(current_frame, cv2.COLOR_BGR2GRAY)
    current_frame = cv2.resize(current_frame,(360,180))
    match_img(current_frame,img1)
    cv2.imshow('frame', current_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
